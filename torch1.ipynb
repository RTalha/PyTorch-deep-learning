#auto grad function in torch


```
x=torch.tensor([7],dtype=torch.float64,requires_grad=True,device=cuda0)
y=torch.tensor([5],dtype=torch.float64,requires_grad=True,device=cuda0)
```

```
z = ((x**4)*(y**2)) + ((x**2)*y)
```

```
total=torch.sum(z)
```

```
total
```

```
total.backward()
```

```
print(x.grad,y.grad)
```
